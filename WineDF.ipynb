# Aplicar PCA

Nesta seguinte celula vamos fazer os imports de tudo o que nos falta para aplicarmos o PCA

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

Nesta celula vamos fazer import do matplotlib para podermos fazer alguns graficos mais a frente

```python
import matplotlib.pyplot as plt
```

Na celula a seguir usamos o sklearn.preprocessing.**scale**para **standardizar** os dados da **var_independente**

```python
var_ind_scaled = preprocessing.scale(var_independente)
```

Agora vamos criar o objeto **PCA**

```python
pca = PCA()
```

Com o objeto **PCA** criado podemos passar ao calculos propriamente ditos

```python
pca.fit(var_ind_scaled)
```

Agora vamos gerar cordenadas para um grafico de PCA

## PCA Ploting

Primeiro vamos gerar coordenadas baseadas nos **loading scores** para o grafico PCA

```python
pca_data = pca.transform(var_ind_scaled)
```

Com isto feito podemos passar a fazer um grafico que nos mostre quanta **variancia** cobre cada uma das **_principal components_**

Na linha das legendas nos criamos legendas começadas por **PC** seguidas de um numero. São criadas quantas legendas quanto o numero de **principal components**. O "+1" adicionado ao tamanho do array de percentagens deve/se ao facto de o **range excluir** o ultimo numero.

```python
percent_var = np.round(pca.explained_variance_ratio_ * 100, decimals = 1)
legendas = ['PC' + str(count) for count in range(1, len(percent_var) + 1)]
```

### Grafico de barras

Agora ja temos tudo o que é preciso para fazer o nosso grafico de barras, so precisamos de o criar

```python
plt.bar(x = range(1,len(percent_var)+1), height = percent_var, tick_label = legendas)
plt.ylabel("Percentagem de variação")
plt.xlabel("Principal component")
plt.show()
```

```python
pca.explained_variance_ratio_
```

#### Observações e comparação com as recomendações

Como podemos ver as componentes pricipais obtidas **descrevem** bastante **menos** do que o seria **otimo**, tal como o que foi previsto no ponto 7. O facto de os dados não serem normais e de existir muito pouca correlação entre os dados faz com que seja dificil omitir variaveis, por cada uma delas descrever algo distinto. 

### Scatter Plot

Apesar das duas primeiras componentes principais descreverem menos de metade da variancia dos dados vamos realizar um scatter plot para efeito de visualização dos dados

```python
pca_df = pd.DataFrame(pca_data, columns=[f"PC{i}" for i in range(1,pca_data.shape[1]+1) ])
```

```python
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.xlabel("PC1 - {0}%".format(percent_var[0]))
plt.ylabel("PC2 - {0}%".format(percent_var[1]))
```

Como podemos ver, o scatter plot apresentanos apenas um unico cluster, o que não se traduz em nenhuma informação adequirida

# Separar dataset entre Training e Testing

Para separar dataset em training e testing sub-datasets vamos usar **sklearn.model_selection.train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
colunas = []

for col in wine_df.columns: 
     colunas.append(col)
    
colunas.remove('quality')

print(colunas)


```

```python
X = wine_df[colunas]
y = wine_df[["quality"]]
y = y.to_numpy()
y = np.ravel(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

# Criação do Modelo de regressão logistica

## Usar partição 75% training e 25%

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

### Precision Score

```python
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_train,y_pred_train))
print(accuracy_score(y_test,y_pred_test))
```

### Recall Score

```python
from sklearn.metrics import recall_score
```

```python
print(recall_score(y_train,y_pred_train))
print(recall_score(y_test,y_pred_test))
```

### Interpretação

Para os valores de **precisão** podemos ver que dos valores que o modelo aponta como verdadeiros, pouco mais de metade estão corretos e visto que de todos os casos **verdadeiros** 
(TP + FN), apenas cerca de metade é que são assinalados corretamente (**Recall = 59%**), achamos que estes resultados nao favorecem muito o modelo

# F1-score e MCC

### MCC

```python
from sklearn.metrics import matthews_corrcoef
```

```python
print(matthews_corrcoef(y_train,y_pred_train))
print(matthews_corrcoef(y_test,y_pred_test))
```

### Interpretação

Visto que o **F1-Score** é traduzido de uma relação entre **precision** e **recall** (sendo este uma media harmoniosa destes os dois). Como calculado a cima podemos ver que não existe uma distribuição desproporcional da variavel de qualidade, logo se o nosso objetivo fosse avaliar o sistema de uma forma geral seria mais indicado usar a metrica de **accuracy** visto que a sua mais intuitiva do que a do **f-value**. Em relação aos valores obtidos podemos ver que obtemos um valor entre os valores de **recall** e **precision** como era de esperar da metrica. para além disso o valor nao difere muito das outras duas metricas visto que a distribuição da variavel qualidade não é desproporcional.

Olhando agora para o **MCC** podemos ver que este apresenta um valor positivo, como era de esperar visto que os valores de **recall** e **precision** eram superirores a 50%.
Com **MCC =~ 0.33** podemos dizer que exite uma correlação moderada entre os valores reais e os valores previstos.

## Avaliação Final

Para concluir podemos dizer dizer que o modelo de uma forma geral, apesar de ser melhor do que escolher ao acaso, tem um desempanho fraco na previsão de casos positivos estando pouco mais de metade dos casos previstos, corretos.

# Cross Validation: k-fold 

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer
```

```python
standardizer = StandardScaler()
logit = LogisticRegression()

pipeline = make_pipeline(standardizer, logit)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1)
```

```python
scorer = make_scorer(matthews_corrcoef)
```

```python
cv_results = cross_val_score(pipeline,X_train,y_train, cv=kf, scoring=scorer,n_jobs=-1)
```

```python
cv_results.mean()
```

#### Interpretação

Olhando para o resutado podemos ver que a media apresenta um valor superiror ao previamente calculado com os dados de treino. Sabendo que os valores dados pelo Cross validation são mais precisos que os valores oferecidos apenas por uma iteração do calculo do MCC, podemos concluir que o modelo é ligeiramente melhor que o anteriormente avaliado, no entanto trata-se de uma diferença quase residual.
